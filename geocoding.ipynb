```python
import geocoder
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('df.csv',index_col=0)
data.head(3)
```

```python
peter = data.city[0]
print(peter)
g = geocoder.yandex(peter)
print(g.ok == True)
x, y = g.current_result.geometry.items()
print(x)
print(y[1][0])
```

```python
cities = pd.unique(data.city)
cities_map = {}
cities_map[peter] = [y[1][0], y[1][1]]
i = 0
fin = len(cities)
for city in cities:    
    print(city, i/fin*100 )
    g = geocoder.yandex(city)
    
    if g.ok == True:
        typ, point = g.current_result.geometry.items()
        cities_map[city] = [point[1][0], point[1][1]]
    else:
        cities_map[city] = [np.nan, np.nan]
    i+=1    
    
print(cities_map)
```

```python
print(len(data))
city_lat = []
city_lon = []

for ind, city in data.city.iteritems():            
    city_lon.append(cities_map[city][0])
    city_lat.append(cities_map[city][1])
    
print(city_lat[0], city_lon[0])
print(len(city_lat), len(city_lon))
```

```python
c_lat = city_lat
c_lon = city_lon
city_lat = pd.Series(city_lat)
city_lon = pd.Series(city_lon)
data['city_lat'] = city_lat
data['city_lon'] = city_lon
data.head()
```
