```python
test_data = pd.read_csv('test_set.csv')
test_data.head()
```

```python
cities_map = np.load('cities_map.npy').item()
cities_map
```

```python
df_t =  test_data.drop(['amount', 'atm_address', 'country', 'currency', 'mcc', 'terminal_id', 'transaction_date', 'pos_address'], axis=1)
df_t.head()
```

```python
df = df_t
df = df.dropna(subset=['city'])
df.city = df.city.str.upper()
df.head()
df.loc[df['city'].str.contains('MOS'), 'city'] = 'MOSCOW'
df.loc[df['city'].str.contains('PETER'), 'city'] = 'ST-PETERSBURG'
df.city = df.city.str.strip()
df.city.value_counts()
```

```python
city_lat = []
city_lon = []
mis_cit = []

for ind, city in df.city.iteritems():   
    if city not in cities_map:
        mis_cit.append(city)
    else:
        city_lon.append(cities_map[city][0])
        city_lat.append(cities_map[city][1])
    
print(city_lat[0], city_lon[0])
print(len(city_lat), len(city_lon))
```

```python
import geocoder
cities = pd.unique(mis_cit)
cities_map_mis = {}
i = 0
fin = len(cities)
for city in cities:    
    print(city, i/fin*100 )
    g = geocoder.yandex(city)    
    if g.ok == True:
        typ, point = g.current_result.geometry.items()
        cities_map_mis[city] = [point[1][0], point[1][1]]
    else:
        cities_map_mis[city] = [np.nan, np.nan]
    i+=1    
    
print(cities_map_mis)
```

```python
spend_lat = []
spend_lon = []

for index, row in df.iterrows():    
    if (pd.notnull(row['pos_address_lat']) and pd.notnull(row['pos_address_lat'])):
        spend_lat.append(row['pos_address_lat'])
        spend_lon.append(row['pos_address_lon'])
    elif pd.notnull(row['pos_address_lat']):
        spend_lat.append(row['pos_address_lat'])
        spend_lon.append(row['pos_address_lon'])
    elif pd.notnull(row['atm_address_lat']):
        spend_lat.append(row['atm_address_lat'])
        spend_lon.append(row['atm_address_lon'])
    else:
        spend_lat.append(np.nan)
        spend_lon.append(np.nan)
    
```

```python
# spend_lat = pd.Series(spend_lat)
# spend_lon = pd.Series(spend_lon)
# print(len(spend_lat))
# print(len(df))

df['spend_lat'] = spend_lat.values
df['spend_lon'] = spend_lon.values
df.head()
```

```python
df = df.drop(['atm_address_lat', 'atm_address_lon', 'pos_address_lon', 'pos_address_lon'], axis=1)
df.head()
```

```python
# print(len(data))
city_lat = []
city_lon = []

for ind, row in df.iterrows():
    city = row['city']        
    if city in cities_map:
        city_lon.append(cities_map[city][0])
        city_lat.append(cities_map[city][1])
    elif city in cities_map_mis:
        city_lon.append(cities_map_mis[city][0])
        city_lat.append(cities_map_mis[city][1])
    else:
        cities_map[city] = [row['spend_lat'],row['spend_lon']]
        
print(city_lat[0], city_lon[0])
print(len(city_lat), len(city_lon))
```

```python
city_lat = pd.Series(city_lat)
city_lon = pd.Series(city_lon)
df['city_lat'] = city_lat.values
df['city_lon'] = city_lon.values
```

```python
np.save('cities_map_mis.npy', cities_map_mis) 
```
